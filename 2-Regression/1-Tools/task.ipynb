## Diabetes

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X_, y = datasets.load_diabetes(return_X_y=True)
print(X_.shape)
print(X_[0])
```

```python
X = X_[:, np.newaxis, 9]
X[:5]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_train, y_train,  color='lightgrey')
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

## Linnerud

```python
# X - 0 Chins, 1 Situps, 2 Jumps
# y - 0 Weight, 1 Waist, 2 Pulse
X_, y_ = datasets.load_linnerud(return_X_y=True)
```

```python
# 1 - situps
X = X_[:, np.newaxis, 2]
y = y_[:, np.newaxis, 1]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_train, y_train,  color='lightgrey')
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Situps')
plt.ylabel('Waistline')
plt.title('Linnerud dataset')
plt.show()
```
